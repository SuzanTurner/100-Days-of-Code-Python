```python
import tkinter
import time

window = tkinter.Tk()
window.minsize(width = 400, height = 400)
window.title("Typing Speed Test")
window.configure(bg = "white")

text = tkinter.Label(text = "Type Out This Message!", bg = "white", font = ("Helvetica", 15, "bold"), highlightthickness = 2, highlightbackground = "black")
text.grid(row = 1, column = 0, padx =  5, pady = 5)

para_text = "The quick brown fox jumps over the lazy dog \n and sits over there while munching snackies \n and giving quickies to whoever tf is in Ramdeobaby. \n And bro that's not it, it's kinda crazy in here. "
para = tkinter.Label()
para.configure(text = para_text, bg = "white", font = ("Helvetica", 12, "normal"), padx = 5, pady = 5)
para.grid(row = 2, column = 0, padx = 5, pady = 5)

def on_key_release(event):
    global start, w
    w = 0
    user_text = t.get("1.0", "end-1c")  
    start = time.time()
    if user_text == para_text[:len(user_text)]:  
        #print("Correct so far!")  
        if user_text == para_text[:len(user_text)] and user_text == " ":
            w += 1
    else:
        print("AYY, Mismatch detected!")  

t = tkinter.Text()
t.configure(width = 50, height = 5, highlightthickness = 1, highlightbackground = "black" )
t.bind("<KeyRelease>", on_key_release)
t.grid(row = 3, column = 0, padx = 5, pady = 5)

def get_wpm():
    global start
    if start is None:
        return  

    end = time.time()
    total_time = (end - start) / 60 
    wpm_ = w / total_time if total_time > 0 else 0
    wpm = tkinter.Label()
    wpm.configure(text = f"WPM is: {wpm_}", font = ("Helvetica", 12, "normal"), padx = 5, pady = 5) 
    wpm.grid(row = 5, column = 0, padx = 5, pady = 5)

send = tkinter.Button(command = get_wpm)
send.configure(text = "Send", font = ("Helvetica", 13, "normal"))
send.grid(row = 4, column = 0, padx = 5, pady = 5)
window.mainloop()
```

```python
import requests
from datetime import datetime
import os

# Personal details
GENDER = "Female"
WEIGHT_KG = 46
HEIGHT_CM = 166
AGE = 19

# API credentials
APP_ID = "3892e346"
API_KEY = "52dccf2543ef1cbdc6eb9362d63d60ce"
SHEET_ENDPOINT = "https://api.sheety.co/a6468c3915ec0f04bd797b009b1e1ba8/copyOfWorkouts2/workouts"
TOKEN = "AdiPadi"

exercise_endpoint = "https://trackapi.nutritionix.com/v2/natural/exercise"

# Headers for Nutritionix API
headers = {
    "x-app-id": APP_ID,
    "x-app-key": API_KEY,
}

# Headers for Sheety API
bearer_headers = {
    "Authorization": f"Bearer {TOKEN}"
}

while True:
    exercise_text = input("Tell me which exercises you did (or type 'exit' to finish): ")

    if exercise_text.lower() == 'exit':
        print("Exiting...")
        break

    # Parameters for the exercise query
    parameters = {
        "query": exercise_text,
        "gender": GENDER,
        "weight_kg": WEIGHT_KG,
        "height_cm": HEIGHT_CM,
        "age": AGE
    }

    # Sending the request to the Nutritionix API
    response = requests.post(exercise_endpoint, json=parameters, headers=headers)
    result = response.json()

    # Date and time for the entry
    today_date = datetime.now().strftime("%d/%m/%Y")
    now_time = datetime.now().strftime("%X")

    # Posting the results to Sheety
    for exercise in result["exercises"]:
        sheet_inputs = {
            "workout": {
                "date": today_date,
                "time": now_time,
                "exercise": exercise["name"].title(),
                "duration": exercise["duration_min"],
                "calories": exercise["nf_calories"]
            }
        }

        sheet_response = requests.post(SHEET_ENDPOINT, json=sheet_inputs, headers=bearer_headers)
        print(sheet_response.text)
```

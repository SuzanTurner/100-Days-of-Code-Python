```python

def d(score):
    while True:
        n = random.randint(0,49)
        dis = data[n]['description']
        place = data[n]['country']
        fol1 = data[n]['follower_count']
        name = data[n]['name']
    
        numbers = list(range(50))
        numbers_except_n= [x for x in numbers if x != n]
        notn = random.choice(numbers_except_n)
    
        dis2 = data[notn]['description']
        place2 = data[notn]['country']
        fol2 = data[notn]['follower_count']
        name2 = data[notn]['name']
        
        print(data[n]['name'], f"\n{dis}\n{place}\nFollowers: {fol2}")
        print(vs)
        print(data[notn]['name'], f"\n{dis2}\n{place2}")
        
        
        clear_output(wait=True)
        
        ch = input(f"Does {name2} have higher or lower followers than {name}? (Higher/Lower): ").lower()
        if (fol1 > fol2 and ch == "lower") or (fol2 > fol1 and ch == "higher"):
            score += 1
            #clear_output(wait=True)
            time.sleep(1)
            print("Correct!!")
            print(f"Your score is: {score}\n")
        else:
            if fol1 > fol2:
                correct_name = name
                print(f"{name2} has fewer followers than {correct_name} \nGame Over")
            else:
                correct_name = name2
                print(f"{name} has fewer followers than {correct_name} \nGame Over")
            break
        


print(logo)
score = 0
print("You must guess who has more followers based on description and country \n")
d(score)
        


```
